```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import timeit
```

```python
total = pd.DataFrame()

for i in range(1, 11):
    source = urllib.urlopen('http://wwe2.osc.state.ny.us/transparency/contracts/contractresults.cfm?PageNum_rsContract='+ str(i) +'&sb=a&a=Z0000&ac=&v=%28Enter+Vendor+Name%29&vo=B&cn=&c=-1&m1=0&y1=0&m2=0&y2=0&am=0&b=Search&order=VENDOR_NAME&sort=ASC')
    soup = BeautifulSoup(source.read())
    table = soup.find('table')
    table_body = table.find('tr')
    colnames = []
    th = soup.find_all('th')
    for name in th:
        colnames.append(name.find_all('a')[1].text)
        
    contract_data = []
    table = soup.find('table')
    rows = table.find_all('tr')

    for row in rows[1:]:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        contract_data.append([ele for ele in cols if ele])

    contracts = pd.DataFrame(contract_data, columns = colnames)
    total = total.append(contracts, ignore_index=True)
```
